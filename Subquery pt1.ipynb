## Example of Subqueries (Fetch Data From table based on another table)

```python
%%sql 
SELECT DISTINCT(department)
FROM employees
```

```python
%%sql
SELECT DISTINCT(department) 
FROM departments
```

### Querying a Department From employees table that does not exist in departments table.

```python
%%sql
SELECT * FROM employees
WHERE department NOT IN (SELECT department FROM departments)
LIMIT 5
```

### Therefore all departements were selected but Camping & Fishing department.

```python
%%sql
SELECT department FROM departments
EXCEPT
SELECT department FROM employees
```

## Another Cases

1. Query Departments & Employee Table, take a note on department table in division column, try to fetch data from employees based information of division column in this case use 'Electronics' division in departments table.

```python
%%sql 
SELECT * FROM EMPLOYEES
LIMIT 5
```

```python
%%sql
SELECT * FROM departments
LIMIT 5
```

```python
%%sql
SELECT department  FROM employees
WHERE department IN (SELECT department FROM departments WHERE division = 'Electronics')
LIMIT 5
```

2. Try to fetch employees that work in country either 'Canada' or 'Asia', use regions table information to fetch the data.

```python
%%sql 
SELECT * FROM regions
LIMIT 5
```

```python
%%sql
SELECT department, region_id
FROM employees
WHERE region_id IN (SELECT region_id FROM regions WHERE country = 'Asia' OR country = 'Canada')
```
