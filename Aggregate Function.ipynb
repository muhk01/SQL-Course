### Excercise 1.

1. Write Quesry Shown States and Largest Amount of Fruit Supply

```python
%%sql
SELECT state, SUM(supply)
FROM fruit_imports
GROUP BY state 
ORDER BY SUM(supply) desc
```

2. Write a query that returns the most expensive cost_per_unit of every season. The query should display 2 columns, the season and the cost_per_unit

```python
%%sql
SELECT season, MAX(cost_per_unit)
FROM fruit_imports
GROUP BY season
ORDER BY MAX(cost_per_unit) desc
```

3. Write a query that returns the state that has more than 1 import of the same fruit.

```python
%%sql
SELECT state, COUNT(name)
FROM fruit_imports
GROUP BY state
HAVING COUNT(name) > 1
ORDER BY COUNT(name) desc
```

4. Write a query that returns the seasons that produce either 3 fruits or 4 fruits.

```python
%%sql
SELECT season, COUNT(*) as kind
FROM fruit_imports
GROUP BY season
HAVING COUNT(*) > 2
```

5. Write a query that takes into consideration the supply and cost_per_unit columns for determining the total cost and returns the most expensive state with the total cost.

```python
%%sql
SELECT state, 
SUM(supply * cost_per_unit) as total_cost
FROM fruit_imports
GROUP BY state
ORDER BY total_cost desc
```

4. Group Salary By Gender and Region

```python
%%sql
SELECT gender, region_id, 
round(AVG(salary)) as avg_sal, 
MIN(salary) as min_sal,
MAX(salary) as max_sal
FROM employees
GROUP BY gender, region_id
ORDER BY gender desc, region_id asc
```
